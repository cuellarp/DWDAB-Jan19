**Working with the Data**

```python
dupe = df.duplicated() #creates list of True/False values
df[dupe] #shows rows where duplicated is True
```

```python
#drops duplicate rows, default is to keep first observation
df.drop_duplicates()
```

```python
#show the column names in the dataset
df.columns
#or print(df.columns)
```

```python
#show the data type of each column
df.dtypes
```

```python
#Creating a new dataframe labeled df2 that only has 5 columns.

df2=df[['FULLADDRESS','AYB','SALEDATE','WARD', 'PRICE', 'STYLE','STRUCT', 'LANDAREA', 'QUADRANT']]
df2.head()
```

```python
missing=df2['FULLADDRESS'].isnull()^df2['AYB'].isnull()^df2['WARD'].isnull()^df2['PRICE'].isnull()^df2['LANDAREA'].isnull()
df2.loc[missing]
```

```python
#set indices as addressesnumbers
df2.set_index('WARD', inplace=True)
df2.head()
```

```python
#Make scatterplott with more details to scatterplot
sns.lmplot(x='PRICE', y='LANDAREA', data=df2, 
           fit_reg=False, #remove regression line
           hue='QUADRANT')   #color by evolution ward
```

```python
#boxplot showing stats for each characteristic
sns.boxplot(data=df2)
```
